```python
# Import modules
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pickle
import matplotlib.pyplot as plt
```

```python
# Importing data
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
# Fitting Simple Linear Regression to the Training set
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
print('linear model intercept: {}'.format(regressor.intercept_))
print('linear model coeff:\n{}'.format(regressor.coef_))
print('R-squared score (training): {:.3f}'.format(regressor.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'.format(regressor.score(X_test, y_test)))

plt.figure(figsize=(5,4))
plt.scatter(X, y, marker= 'o', s=50, alpha=0.8)
plt.plot(X, regressor.coef_ * X + regressor.intercept_, 'r-')
plt.title('Least-squares linear regression')
plt.xlabel('Feature value (X)')
plt.ylabel('Target value (y)')
plt.show()
```

```python
# Predicting the Test set results
y_pred = regressor.predict(X_test)
```

```python
# Saving model to disk
pickle.dump(regressor, open('model.pkl','wb'))
```

```python
# Loading model to compare the results
model = pickle.load(open('model.pkl','rb'))
print(model.predict([[1.8]]))
```
